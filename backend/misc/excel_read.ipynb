```python
import numpy as np
import pandas as pd
```

```python
table = pd.read_excel("BSEMC curriculum 2024.xlsx", header=None)
table
```

```python
table.dropna(subset=[0], inplace=True)
table
```

```python
#table.drop(columns=["Unnamed: 2", "Unnamed: 3", "Unnamed: 5", "Unnamed: 8"], inplace=True)
table
```

```python
table.rename(columns={0:"Course Code", 1: "Course Name", 2: "Hours_Lec",
                      3: "Hours_Lab", 4: "Units_Lec", 5: "Units_Lab",
                      6: "Pre-requisite"}, inplace=True)
table
```

```python
table = table[(table["Course Code"] != "TOTAL") & 
              (table["Course Code"] != "TOTAL ACADEMIC UNIT(S):") & 
              (table["Course Code"] != "Course Code")]
table
```

```python
table["Hours"] = np.where(table["Hours_Lec"] > 0, table["Hours_Lec"], table["Hours_Lab"])
table["Units"] = np.where(table["Units_Lec"] > 0, table["Units_Lec"], table["Units_Lab"])
table
```

```python
table.drop(columns=["Units_Lec", "Units_Lab", "Hours_Lec", "Hours_Lab"], inplace=True)
table
```

```python
table["Year"] = table["Course Code"].where(table["Course Code"].str.contains("Year"))
table["Year"].ffill(inplace=True)
table = table[~(table["Course Code"].str.contains("Year"))].reset_index(drop=True)
table
```

```python
table["Semester"] = table["Course Code"].where(table["Course Code"].str.contains("Semester"))
table["Semester"].ffill(inplace=True)
table = table[~(table["Course Code"].str.contains("Semester"))].reset_index(drop=True)
table
```

```python
table.replace(["First Semester", "Second Semester", "First Year", "Second Year", "Third Year ", "Fourth Year "], [1, 2, 1, 2, 3, 4], inplace=True)
table
```

```python
table.to_csv('BSEMC.csv', index=False)
```
